```python
%matplotlib inline
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
#El experimento se trata de encontrar el coeficiente de expansion lineal del aluminio, para el experimento se cuenta con 
#una varilla hueca a temperatura ambiente con una longitud de 30cm, se mide la expansion con un reloj comparador, al hacer
#pasar vapor de agua por la varilla. En este experimento se esta considerando el doble de la incertidumbre de los 
#instrumentos ya que se esta considerando el error humano tambien (El experimento esta basado en el laboratorio 3 del  
#curso de fisica experimental 2.
DT=56.0
DX=0.04
Xo=30.0
alfa=DX/(Xo*DT)
p=[]
x=[]
minimos=[]
maximos=[]
medias=[]
for i in range(100):
    x.append(i)
    p_t=np.zeros(i+2)
    for j in range(i+2):
        alfa1=np.zeros(100)
        for k in range(100):
            dT1=DT+np.random.random_integers(-2,2)
            dx1=DX+0.001*np.random.random_integers(-2,2)
            xo1=Xo+0.1*np.random.random_integers(-2,2)
            alfa1[k]=dx1/(dT1*xo1)
        p_t[j]=np.average(alfa1)
    ds=np.std(p_t)
    av=np.average(p_t)
    t=(av-alfa)/ds 
    p.append(1-stats.t.sf(np.abs(t), 1)*2)
    minimos.append(min(p))
    maximos.append(max(p))
    medias.append(np.average(p))
```

```python
w=plt.figure(1)
plt.scatter(x,p)
plt.title('p-values en varias repeticiones del experimento')
plt.xlabel('Numero de experimentos')
plt.ylabel('p-value')
w.show()
```

```python
q=plt.figure(2)
plt.scatter(x,minimos)
plt.title('Minimos de p-value')
plt.xlabel('Numero de experimentos')
plt.ylabel('Min p-value')
q.show()
```

```python
e=plt.figure(3)
plt.scatter(x,maximos)
plt.title('Maximos de p-value')
plt.xlabel('Numero de experimentos')
plt.ylabel('Max p-value')
e.show()
```

```python
y=plt.figure(4)
plt.scatter(x,medias)
plt.title('Medias de p-value')
plt.xlabel('Numero de experimentos')
plt.ylabel('Media p-value')
y.show()
```
